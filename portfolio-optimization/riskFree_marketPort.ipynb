**Objective:** Construct an optimal portfolio by combining a risk-free asset and the market portfolio to achieve a target standard deviation of 8%. We'll calculate the expected return, portfolio weights, and visualize the efficient frontier.


```python
#Defining given parameters and assumptions for calculations 

risk_free_rate = 0.04 #risk free rate of 4%
market_return = 0.10 #expected market return of 10%
market_std_dev = 0.20 #market standard deviation of 20%
target_std_dev = 0.08 #target portfolio deviation of 8%
```

**Risk-Free rate**: this is the return of a risk free asset typically a government bond 
**Market return**: expected return of the market portfolio 
**Market standard deviation**: volatility of the risk associated with the market portfolio 
**target standard deviation**: desired volatility level for optimal portfolio

```python
#Calculating market risk premium and sharpe ratio 

# The market risk premium is the additional return expected from holding a risky market
# portfolio instead of risk-free assets. 

#The sharpe ratio measures the risk-adjusted return of an investment

market_risk_premium = market_return - risk_free_rate
market_sharpe_ratio = market_risk_premium / market_std_dev
```

**Market Risk Premium is:** 
- Market Return - Risk free rate 

**Market Sharpe Ratio is:** 
- Market Risk Premium / Market Standard Deviation 

```python
# Determining optimal portfolio weights

# Here we want to find the weights of the risk-free asset and the market portfolio that 
# achieve our target standard deviation 

weight_risky = target_std_dev / market_std_dev
weight_risk_free = 1 - weight_risky
```

**Weight in Risky Asset:**
- Target Standard Deviation / Market Standard deviation

**Weight in Risk-Free Asset:**
- 1 - the weight in risky asset


This works because combining the risk-free asset and the market portfolio linearly, the standard deviation of the portfolio is directly proportional to the weight of the risky asset because the risk-free asset has zero standard deviation 

```python
# Calculating expected return of optimal portfolio 
expected_return = risk_free_rate + weight_risky * market_risk_premium
```

**Expected Return:**
- Risk free rate + Weight of the risk * the additional retunr expected from holding a risky market e.g. market risk premium

```python
# Print the results
print(f"Optimal Portfolio Composition:")
print(f"Weight in Market Portfolio: {weight_risky:.2%}")
print(f"Weight in Risk-Free Asset: {weight_risk_free:.2%}")
print(f"Expected Return of Optimal Portfolio: {expected_return:.2%}")
```

- **Weight in Market Portfolio:** 40%
- **Weight in Risk-Free Asset:** 60%
- **Expected Return:** 6.40%


This means that to achieve a portfolio with 8% standard deviation, one should invest 40% in the market portfolio and 60% in the risk-free asset. 

The expected return of this portfolio would be 6.40%

```python
# Visualizing the efficient frontier and optimal portfolio
import numpy as np
import matplotlib.pyplot as plt

def calculate_portfolio_return(weight):
    return risk_free_rate + weight * market_risk_premium

def calculate_portfolio_std_dev(weight):
    return weight * market_std_dev

weights = np.linspace(0, 2, 200)  # Allows for leveraging (weights > 1)
returns = [calculate_portfolio_return(w) for w in weights]
std_devs = [calculate_portfolio_std_dev(w) for w in weights]

plt.figure(figsize=(12, 8))
plt.plot(std_devs, returns, 'b-', label='Capital Market Line')
plt.plot(market_std_dev, market_return, 'ro', markersize=8, label='Market Portfolio')
plt.plot(0, risk_free_rate, 'go', markersize=8, label='Risk-Free Asset')
plt.plot(target_std_dev, expected_return, 'yo', markersize=8, label='Optimal Portfolio')
plt.xlabel('Standard Deviation (Risk)')
plt.ylabel('Expected Return')
plt.title('Capital Market Line and Optimal Portfolio')
plt.legend()
plt.grid(True)
plt.show()
```

**Capital Market Line (CML)**: 
- This is the straight line that connects the risk-free rate and the market portfolio on the risk-return plot

**Weights**: 
- These range from 0 to 2 to show the effect of borrowing aka leverage, when weight exceed 1.

#### Elements of the plot 
- **Blue Line:** Represents the CML, showing the trade-off between risk and return.
- **Red Dot:** Market portfolio.
- **Green Dot:** Risk-free asset.
- **Yellow Dot:** Optimal portfolio with target standard deviation.


In all, this plot shows that by moving along the CML, one can achieve different combinations of risk and return by adjusting the weights betweent the risk-free asset and the market portfolio

```python
# Calulating sharpe ration of optimal portfolio 

optimal_sharpe_ratio = (expected_return - risk_free_rate) / target_std_dev
print(f"Sharpe Ratio of Optimal Portfolio: {optimal_sharpe_ratio:.4f}")
print(f"Market Sharpe Ratio: {market_sharpe_ratio:.4f}")
```

The sharpe ration of the optimal portfolio equals that of the market portfolio which is what is expected when combining a risk-free asset with the market portfolio along the CML.
